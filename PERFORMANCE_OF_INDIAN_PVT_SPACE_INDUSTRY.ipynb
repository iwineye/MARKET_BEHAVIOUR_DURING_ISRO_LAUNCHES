# MAXIMUM RETURNS

***We'll see the maximum returns a person would get if they held equal number of stocks and sold them at right time.***

```python
#Adding the columns of profir and value of the stock and calculating them for each day

company_list = [BHEL,CENTUM,GODREJIND,HAL,LT,MTARTECH,PARAS,TATAELXSI]
company_name = ['BHEL.LTD','CENTUM.LTD','GODREJ.LTD','HAL.LTD','L&T.LTD','MTARTECH.LTD','PARAS.LTD','TATAELXSI.LTD']

for company, com_name in zip(company_list, company_name):
    company['PROFIT']=company['Daily Return']*int(stocks_holding)
    company['STOCK_VALUE']=company['Adj Close']*int(stocks_holding)
    company["company_name"]=com_name

    max_returns= pd.concat(company_list, axis=0)
    max_returns=max_returns.drop(['Unnamed: 0'],axis=1)
    max_returns['Date']=pd.to_datetime(max_returns['Date'])
    #max_returns=max_returns.set_index('Date')
```

```python
max_returns
```

```python
#creating the data of max returns
idx = max_returns.groupby('company_name')['STOCK_VALUE'].idxmax()
max_scores = max_returns.loc[idx]
max_scores.groupby('company_name')
max_scores.head(25)
```

```python
#Creating a function to pick the row with maximum profit of each company

def get_max_score(group):
    return group.loc[group['PROFIT'].idxmax()]

max_scores = max_returns.groupby('company_name').apply(get_max_score)
```

```python
profits=pd.DataFrame(max_scores,columns=['Date','High','Adj Close','Volume','PROFIT','STOCK_VALUE'])
```

```python
profits
```

# EXPLONENTIAL MOVING AVERAGE

***Unlike moving average it takes only recent data into consideration thus avoid the lag factor. Thus making EMA one of the reliable metrics to people doing short term trading for evaluating their positions.***

```python
# comparing exponential moving average with simple moving average
BHEL['ema'] = BHEL['Close'].ewm(span=3, adjust=False, min_periods=5).mean()
#print(BHEL)
```

```python
df3.info()
```

```python
plt.figure(figsize=(15, 10))
#BHEL[].plot()
BHEL[['Adj Close','ema','MA for 5 days','MA for 8 days','MA for 13 days']].plot()
#axes[0,0].set_title('BHEL.LTD')
plt.ylabel('ema')
plt.xlabel(None)
plt.title(f"BHEL")
```

**EMA is often analysed with adjusted close as a line of reference.**

***EMA is often calculated and compared between the periods of 5-8-13 days to monitor a short term changes. This plays a key role in deciding when to sell and hold a stock***

**bold text**

```python
ema_day = [5,8,13]

for ema in ema_day:
    for company in data:
        column_name = f"EMA for {ema} days"
        company[column_name] = company['Close'].ewm(span=ema, adjust=False, min_periods=5).mean()


fig, axes = plt.subplots(nrows=4, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

#BHEL,CENTUM,GODREJIND,HAL,LT,MTARTECH,PARAS,TATAELXSI

BHEL[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[0,0])
axes[0,0].set_title('BHEL.LTD')

CENTUM[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[0,1])
axes[0,1].set_title('CENTUM.LTD')

GODREJIND[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[1,0])
axes[1,0].set_title('GODREJ.LTD')

HAL[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[1,1])
axes[1,1].set_title('HAL.LTD')

LT[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[2,0])
axes[2,0].set_title('L&T')

MTARTECH[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[2,1])
axes[2,1].set_title('MTARTECH.LTD')

PARAS[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[3,0])
axes[3,0].set_title('PARAS.LTD')

TATAELXSI[['EMA for 5 days','EMA for 8 days','EMA for 13 days']].plot(ax=axes[3,1])
axes[3,1].set_title('TATAELXSI.LTD')

fig.tight_layout()
```

# PERFORMANCE ANALYSIS

***Bullish Signal*** :When the 5 EMA crosses above the 8 EMA, and both are above the 13 EMA, you can take this as a potential bullish signal.
Consider the 5 EMA as the trigger, the 8 EMA as the intermediate measure, and the 13 EMA as the baseline. The crossover suggests the momentum is in favour of buyers.

***Bearish Signal***:A bearish setup when the 5 EMA crosses below the 8 EMA and both are below the 13 EMA. Sellers are gaining the upper hand, and the opportunity to go short might be present.
***

***From the above note we can see that all the stocks showed a bullish trend through out the time period.***

Calculating the growth in the stock price of each company from the begining of the time period and the closing period of the last date in the *data*.
